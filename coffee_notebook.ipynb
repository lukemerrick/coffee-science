```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Coffee Tasting.csv')
```

```python
thtemp = df[df['Coffee Name'].isin(['190', '200', '212'])]
```

```python
grp = thtemp.groupby('Coffee Name')
```

```python
means = grp.mean()
means
```

```python
features = []
pal = sns.color_palette("hls", thtemp['Coffee Name'].unique().shape[0])
fig = plt.figure(figsize=(12,2))
for i, feature in enumerate(features):
    p = fig.add_subplot(1,5,i+1)
    for cid, coffee in enumerate(thtemp['Coffee Name'].unique()):
        subset = thtemp[thtemp['Coffee Name'] == coffee]
        sns.distplot(subset[feature],
                     #kde = False,
                     norm_hist = True,
                     label = coffee,
                     axlabel = feature[:10],
                     color = pal[cid]
                    )
        p.legend()
plt.show()
```
